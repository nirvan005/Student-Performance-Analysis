<h1>One Hot Encoding</h1>

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#['school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu',
       # 'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime',
       # 'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
       # 'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout', 'Dalc',
       # 'Walc', 'health', 'absences', 'G1', 'G2', 'G3']
```

```python
encode_category=ColumnTransformer([
    ('encode_all',OneHotEncoder(dtype='int8',drop='first',sparse_output=False),[0,1,3,4,5,8,9,10,11,15,16,17,18,19,20,21,22])
],remainder='passthrough')
encode_category.set_output(transform='pandas')
```

```python
X_train_g3_encoded=encode_category.fit_transform(X_train_g3,y_train_g3)
X_train_g3_encoded.sample(5)
```

```python
X_test_g3_encoded.shape
```

```python
X_test_g3_encoded=encode_category.transform(X_test_g3)
```

```python
from sklearn.preprocessing import PowerTransformer
trf2=ColumnTransformer([
    ('yeoTransform',PowerTransformer(),slice(26,41))
],remainder='passthrough')
trf2.set_output(transform='pandas')
X_train_g3_encoded_trf=trf2.fit_transform(X_train_g3_encoded)
X_test_g3_encoded_trf=trf2.transform(X_test_g3_encoded)
```

<h4>Initially I did not consider G1 and G2 considering it a data leakage. I presumed that G3 is sum of G1 and G2 and some other marks.</h4>
<h4>But later these will be considered and G3 is independent of G1 and G2</h4>

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train_g3_encoded_trf,y_train_g3)
```

```python
pred_g3_all=lr.predict(X_test_g3_encoded_trf)
```

```python
X_g3_all=np.concatenate((X_train_g3_encoded_trf,X_test_g3_encoded_trf),axis=0)
y_g3_all=np.concatenate((y_train_g3,y_test_g3),axis=0)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.model_selection import cross_val_score
print("Mean absolute error:",np.mean(cross_val_score(lr,X_g3_all,y_g3_all,cv=10,scoring='neg_mean_absolute_error'))*-1)
print("Mean squared error:",np.mean(cross_val_score(lr,X_g3_all,y_g3_all,cv=10,scoring='neg_mean_squared_error'))*-1)
print("r2 score:",np.mean(cross_val_score(lr,X_g3_all,y_g3_all,cv=10,scoring='r2')))
```

<h1>Final pipeline</h1>

```python
encode=ColumnTransformer([
    ('encode_all',OneHotEncoder(dtype='int8',drop='first',sparse_output=False),[0,1,3,4,5,8,9,10,11,15,16,17,18,19,20,21,22])
],remainder='passthrough')
```

```python
encode.fit_transform(X_train)[:,38]
```

<h4>Absences column can be transformed to make it normalized</h4>

```python
from sklearn.preprocessing import FunctionTransformer
def apply_log1p(x):
    return np.log1p(x.astype(float))
logTransform=ColumnTransformer([
    ('transform_absences',FunctionTransformer(func=apply_log1p),[38])
],remainder='passthrough')
```

```python
actual_model=LinearRegression()
```

```python
performance_predict_pipeline=make_pipeline(encode,logTransform,actual_model)
```

```python
performance_predict_pipeline.fit(X_train,y_train)
```

```python
prediction=performance_predict_pipeline.predict(X_test)
```

```python
print(mean_absolute_error(y_test,prediction))
print(mean_squared_error(y_test,prediction))
print(r2_score(y_test,prediction))
```
